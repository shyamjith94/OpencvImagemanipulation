```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
my_image_1 = "lenna.png"
my_image_2 = "baboon.png"
my_image_3 = "barbara.png"
```

##Flip Image


```python
my_image = cv2.imread(my_image_1)
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(my_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
width, height, c = my_image.shape
print(f"width {width} height {height} c {c}")
```

```python
array_flip = np.zeros((width, height, c), dtype=np.uint8)
```

```python
for i,row in enumerate(my_image):
        array_flip[width-1-i,:,:]=row
```

```python
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(array_flip, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
for flipcode in [0,1,-1]:
    im_flip =  cv2.flip(my_image,flipcode )
    plt.imshow(cv2.cvtColor(im_flip,cv2.COLOR_BGR2RGB))
    plt.title("flipcode: "+str(flipcode))
    plt.show()
```

```python
im_flip = cv2.rotate(my_image,0)
plt.imshow(cv2.cvtColor(im_flip,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
flip = {"ROTATE_90_CLOCKWISE":cv2.ROTATE_90_CLOCKWISE,"ROTATE_90_COUNTERCLOCKWISE":cv2.ROTATE_90_COUNTERCLOCKWISE,"ROTATE_180":cv2.ROTATE_180}
```

```python
for key, value in flip.items():
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(my_image, cv2.COLOR_BGR2RGB))
    plt.title("orignal")
    
    plt.subplot(1,2,2)
    plt.imshow(cv2.cvtColor(cv2.rotate(my_image,value), cv2.COLOR_BGR2RGB))
    plt.title(key)
    plt.show()
```

##Croping Image

```python
upper = 150
lower = 400
crop_top = my_image[upper: lower,:,:]
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(crop_top, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
left = 150
right = 400
crop_horizontal = crop_top[: ,left:right,:]
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(crop_horizontal, cv2.COLOR_BGR2RGB))
plt.show()
```

##Changing Specific Image Pixels

```python
array_sq = np.copy(my_image)
array_sq[upper:lower,left:right,:] = 0
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(my_image,cv2.COLOR_BGR2RGB))
plt.title("orignal")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(array_sq,cv2.COLOR_BGR2RGB))
plt.title("Altered Image")
plt.show()
```

```python
start_point, end_point = (left, upper),(right, lower)
image_draw = np.copy(my_image)

cv2.rectangle(image_draw, pt1=start_point, pt2=end_point, color=(0, 255, 0), thickness=10) 
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(image_draw, cv2.COLOR_BGR2RGB))
plt.show()
```
