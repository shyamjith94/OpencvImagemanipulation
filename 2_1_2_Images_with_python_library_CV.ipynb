```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
my_image_1 = "lenna.png"
my_image_2 = "baboon.png"
my_image_3 = "barbara.png"
```

```python
def get_concat(img1, img2):
    dst = Image.new('RGB', (img1.width + img2.width, img1.height))
    dst.paste(img1, (0, 0))
    dst.paste(img2, (img1.width, 0))
    return dst
```

```python
my_image = cv2.imread(my_image_1)
```

```python
my_image.shape
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(my_image)
plt.show()
```

##Gray Scale Image

```python
my_image_gry = cv2.cvtColor(my_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(my_image_gry)
plt.show()
```

```python
my_image_gry.shape
```

##Color Chanel

```python
blue, green, red = my_image[:, :, 0], my_image[:, :, 1], my_image[:, :, 2]
```

```python
# concat image chanel with image
img_bgr = cv2.vconcat([blue, green, red])
```

```python
plt.figure(figsize=(5,5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(my_image, cv2.COLOR_BGR2GRAY))
plt.title("RGB Image")

plt.subplot(122)
plt.imshow(img_bgr, cmap='gray')
plt.title("Different color channels  blue (top), green (middle), red (bottom)")
plt.show()
```

```python
#we create a new array baboon_red and set all but the red color channels to zero. Therefore, when we display the image, it appears red
baboon_red = my_image.copy()
baboon_red[:, :, 0] = 0
baboon_red[:, :, 1] = 0
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(baboon_red, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
baboon_blue = my_image.copy()
baboon_blue[:, :, 1] = 0
baboon_blue[:, :, 2] = 0
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(baboon_blue, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
baboon_green = my_image.copy()
baboon_green[:, :, 0] = 0
baboon_green[:, :, 2] = 0
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(baboon_green, cv2.COLOR_BGR2RGB))
plt.show()
```
