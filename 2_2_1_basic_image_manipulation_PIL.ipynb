```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
my_image_1 = "lenna.png"
my_image_2 = "baboon.png"
my_image_3 = "barbara.png"
```

##Flip Image

```python
my_image = Image.open(my_image_1)
plt.figure(figsize=(5,5))
plt.imshow(my_image)
```

```python
img_ary = np.array(my_image)
width, height, c = img_ary.shape
print(f"width {width} height {height} c {c}")
```

```python
array_flip = np.zeros((width, height, c), dtype=np.uint8)
```

```python
for i, row in enumerate(img_ary):
    array_flip[width-1-i, :, :] = row
```

```python
plt.figure(figsize=(5,5))
plt.imshow(array_flip)
plt.show()
```

```python
im_flip = ImageOps.flip(my_image)
plt.figure(figsize=(5,5))
plt.imshow(im_flip)
plt.show()
```

```python
im_mirror = ImageOps.mirror(my_image)
plt.figure(figsize=(5,5))
plt.imshow(im_mirror)
plt.show()
```

```python
im_flip = my_image.transpose(1)
plt.figure(figsize=(5,5))
plt.imshow(im_flip)
plt.show()
```

```python
flip = {"FLIP_LEFT_RIGHT": Image.FLIP_LEFT_RIGHT,
        "FLIP_TOP_BOTTOM": Image.FLIP_TOP_BOTTOM,
        "ROTATE_90": Image.ROTATE_90,
        "ROTATE_180": Image.ROTATE_180,
        "ROTATE_270": Image.ROTATE_270,
        "TRANSPOSE": Image.TRANSPOSE, 
        "TRANSVERSE": Image.TRANSVERSE}
```

```python
for key, value in flip.items():
    plt.figure(figsize=(5,5))
    plt.subplot(1,2,1)
    plt.imshow(my_image)
    plt.title("Original")

    plt.subplot(1,2,2)
    plt.imshow(my_image.transpose(value))
    plt.title(key)
    plt.show()
```

##Croping Image

```python
upper = 150
lower = 400
crop_top = img_ary[upper: lower,:,:]
plt.figure(figsize=(5,5))
plt.imshow(crop_top)
plt.show()
```

```python
left = 150
right = 400
crop_horizontal = crop_top[: ,left:right,:]
plt.figure(figsize=(5,5))
plt.imshow(crop_horizontal)
plt.show()
```

```python
image = Image.open(my_image_3)
crop_image = image.crop((left, upper, right, lower))
plt.figure(figsize=(5,5))
plt.imshow(crop_image)
plt.show()
```

##Changing Specific Image Pixels

```python
array_sq = np.copy(img_ary)
array_sq[upper:lower, left:right, 1:2] = 0
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.imshow(img_ary)
plt.title("orignal")

plt.subplot(1,2,2)
plt.imshow(array_sq)
plt.title("Altered Image")
plt.show()
```
